<!-- 파이썬으로 네이버 증권 주요뉴스와 제목, url을 모두 추출하기
제목변수, url변수 만들기 -->

```python

# 주요 뉴스 섹션의 뉴스 항목 선택
srclist = soup.select('.section_strategy a')

# 뉴스 제목과 URL을 저장할 리스트 초기화
news_title = []
news_url = []

# 뉴스 항목을 순회하며 제목과 URL 추출
for item in srclist:
    title = item.get_text(strip=True)
    link = item.get('href')
    if link.startswith('/'):
        link = 'https://finance.naver.com' + link
    
    titles.append(title)
    urls.append(link)

# 결과 출력
for title, url in zip(titles, urls):
    print(f"Title: {title}\nURL: {url}\n")

# 결과를 별도 변수로 저장
title_variable = titles
url_variable = urls

# 저장된 변수 출력
print("Titles:", title_variable)
print("URLs:", url_variable)
```

데이터 프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

```python
print(df)
```

```python
# 곡이름 찾기
len(soup.select('.ellipsis.rank01 a'))
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')]
song
```

```python
# 가수 이름 찾기
# ellpsis rank02로 하면 안됨, 리스트 길이 안맞음
# 가수가 2명이상이거나 동일한 자식속성이 하나 더 있음
artist = []
for item in soup.select('.ellipsis.rank02 a') : 
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
name = [item.text for item in soup.select('.ellipsis.rank02>a')]
len(name)
```

```python
artist = []
for item in soup.select('.checkEllipsis') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
# 인덱스 만들기
rank_list = list(range(1, 101))
rank_list
```

```python
# 사실 판다스에서 인덱스 핸들링 가능함, 다음에 알아보자
songDF = pd.DataFrame({
        "순위":rank_list,
        "노래제목":song,
        "가수명":artist})
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```

## REST Test : GET POST 비교

```python
import requests
```

```python
param = {'name':'Amy', 'age':20, 'address':'Seoul'}
```

```python
# get 방식 - 요청 정보(파라미터)가 인자에 포함되어 있다.
resp1 = requests.get('http://httpbin.org/get', params=param)
print(resp1)
print(resp1.text)
```

```python
# post 방식 - 요청 정보(파라미터)가 인자에 없고, form에 숨겨져있다.(예시라서 보이는 것)
resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2.text)
```
